# 1) BFS for determining the vertex' connected component

```python
Gk = graph.Graph({'node_color': 'yellow', 'font_color': 'b', 'node_size' : 300, 'font_size': 10})
Gk.add_v('A', 2, 3, 4, 5, 6, 7, 8, 99)
Gk.add_adj({'A':(2, 4), 2:('A',4,3,5), 3:(2, 5), 4: ('A', 2, 5, 6), 5:(3, 2, 4, 7), 6:(4, 7), 7: (6, 5), 8:99})
```

```python
Gk.draw()
```

```python
print( graph.bfs_components(Gk, 99) )
print( graph.bfs_components(Gk, 'A') )
```

# 2) BFS for determining a bipartite graph

```python
Gb = graph.Graph({'node_color': 'yellow', 'font_color': 'b', 'node_size' : 300, 'font_size': 10})
Gb.add_v(0, 1, 2, 3, 4, 5)
Gb.add_adj({0:(2,3,4), 1:(2,3,4), 5:(2,4)})
```

```python
Gb.draw()
```

```python
graph.bipartite(Gb)
```

# 3) Euler path in the graph

```python
Ge = graph.Graph({'node_color': 'yellow', 'font_color': 'b', 'node_size' : 300, 'font_size': 10})
Ge.add_v('A', 2, 3, 4, 5, 6, 7)
Ge.add_adj({'A':(2, 4), 2:('A',4,3,5), 3:(2, 5), 4: ('A', 2, 5, 6), 5:(3, 2, 4, 7), 6:(4, 7), 7: (6, 5)})
```

```python
Ge.draw()
```

```python
graph.get_Euler_path(Ge, 5)
```
