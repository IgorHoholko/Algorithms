## BFS for determining the vertex's connected component

```python
Gk = Graph(params_drawing={'node_color': 'green', 'font_color': 'w', 'node_size' : 300, 'font_size': 10})
Gk.add_v('A', 2, 3, 4, 5, 6, 7, 8, 99)
Gk.add_adj({'A':(2, 4), 2:('A',4,3,5), 3:(2, 5), 4: ('A', 2, 5, 6), 5:(3, 2, 4, 7), 6:(4, 7), 7: (6, 5), 8:99})

Gk.draw()
```

# Weighted graph

```python
from graph import utils_weight, generate
```

```python
G = graph.Graph()
G.add_v('A', 'B', 'C', 'D', 'E', 'F', 'G')
G.add_adj({'A':('G', 'B'), 'E':('G','D'), 'C':('D', 'B'), 'F':('A', 'B', 'C', 'D', 'E', 'F')})

G.set_weights([('A', 'G'), ('A', 'B')], [23, 20])
G.set_weights([('F', 'A'), ('F','B'), ('F', 'C'), ('F','D'), ('F', 'E'), ('F', 'G')], [1, 4, 9, 16, 25, 36])
G.set_weights([('E', 'G'), ('E', 'D')], [27, 17])
G.set_weights([('C', 'D'), ('C', 'B')], [3, 15])

G.draw()
```

```python
G.print()
G.print_matrix()
```

## Prime algorithm (build a Minimum Spanning Tree)
<img src="https://upload.wikimedia.org/wikipedia/commons/9/9b/PrimAlgDemo.gif">


```python
MST_p, W_p = utils_weight.get_min_weight_spanner(G)
MST_p.draw()
```

```python
print( "Weight: ", W_p)
MST_p.print_matrix()
```

## Kruskal algorithm (build a Minimum Spanning Tree)
![Kruskal](https://upload.wikimedia.org/wikipedia/commons/b/bb/KruskalDemo.gif)

```python
MST_k, W_k = utils_weight.get_min_weight_spanner(G, mod='K')
MST_k.draw()
```

```python
print( "Weight: ", W_k)
MST_k.print_matrix()
```
