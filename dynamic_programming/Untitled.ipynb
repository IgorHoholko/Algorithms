```python
import numpy as np
import pandas as pd
```

```python
def horse_desk(i, j):
    n, m = i +1, j + 1
    A = np.zeros((n, m), dtype=np.int64)
    A[0,0] = 1
    step_right, step_down = (1, 2), (2, 1)
    
    for i in range(n):
        for j in range(m):
            try:
                A[i + step_right[0], j + step_right[1]] += 2 * A[i, j]  
            except: pass
            try:
                A[i + step_down[0], j + step_down[1]] += 2 * A[i, j]
            except: pass
                
    return pd.DataFrame(A)
```
